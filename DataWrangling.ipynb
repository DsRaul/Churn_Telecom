```python
import pandas as pd
import numpy as np
import requests
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Exportamos los datos

URL = 'https://raw.githubusercontent.com/DsRaul/Churn_Telecom/master/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(URL)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
df.columns[df.isnull().any()]
```

```python
data_1 = df.drop(['customerID'], axis=1)
```

```python
data_1['Churn'].replace(to_replace='Yes', value=1, inplace=True)
data_1['Churn'].replace(to_replace='No',  value=0, inplace=True)
data_1.head()
```

```python
#Let's convert all the categorical variables into dummy variables
df = pd.get_dummies(data_1, columns=['Partner','Dependents','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling'])
df.head()
```
